```python
def extract_responsible_locations_chat_api(descriptions):

    try:
        prompt = (
            "Extract the responsible location from each of the following descriptions. "
            "Provide the results as 'Description {number}: {location}'.\n\n"
            + "\n".join([f"Description {i+1}: {desc}" for i, desc in enumerate(descriptions)])
        )
        
        response = openai.chat.completions.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You are an assistant that extracts locations from descriptions."},
                {"role": "user", "content": prompt},
            ],
            max_tokens=2000,
            temperature=0
        )

        results = response.choices[0].message.content.strip().split("\n")
        extracted_locations = {}
        for result in results:
            if result.startswith("Description"):
                try:
                    desc_number, location = result.split(": ", 1)
                    extracted_locations[desc_number] = location.strip()
                except ValueError:
                    print(f"Error parsing result: {result}")
                    continue

        locations = [
            extracted_locations.get(f"Description {i+1}", "Location not found")
            for i in range(len(descriptions))
        ]

        return locations

    except Exception as e:
        print(f"Error calling OpenAI API: {e}")
        return ["Error" for _ in descriptions]

descriptions = df['description'].tolist()

responsible_locations = extract_responsible_locations_chat_api(descriptions)

df['responsible_location'] = responsible_locations
df.head()
```
